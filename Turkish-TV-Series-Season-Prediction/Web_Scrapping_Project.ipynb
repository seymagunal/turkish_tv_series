```python
import requests
from bs4 import BeautifulSoup


imdb_url = "https://www.imdb.com/search/title/?title_type=tv_series&countries=tr&start=1&ref_=adv_nxt&count=250"

names = [] 
date_fulls = []
times = []
categories = []
ratings = []
stars = []
episodes = []
seasons = []
directeds = []
scriptwriters = [] # writing_credits
release_years = []
production_companies = []
publisher_organization = [] # distrubutors
musics = []
created_bys = [] # Series Produced by 
exported_countries = [] # release_dates

starter = 1
sayi=1

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.84 Safari/537.36',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
}

while starter < 2400:
    if starter != 1:
        imdb_url = f"https://www.imdb.com/search/title/?title_type=tv_series&countries=tr&start={starter}&ref_=adv_nxt&count=250"

    source = requests.get(imdb_url, headers = headers)
    soup = BeautifulSoup(source.text, "html.parser")
    movies = soup.find_all("div", class_ = "lister-item mode-advanced")

    for i in movies: 
        try:
          print(str(sayi))
          sayi=sayi+1
          general_div = i.find("div", class_ = "lister-item-content")
          first_div = general_div.find("h3", class_ = "lister-item-header")

          names.append(first_div.find("a").text if first_div.find("a") else "")

          date_fulls.append(first_div.find("span", class_ = "lister-item-year text-muted unbold").text.strip("()") if first_div.find("span", class_ = "lister-item-year text-muted unbold") else "")

          times.append(general_div.find("span", class_ = "runtime").text.strip() if general_div.find("span", class_ = "runtime") else "")

          categories.append(general_div.find("span", class_ = "genre").text.strip() if general_div.find("span", class_ = "genre") else "")

          ratings.append(general_div.find("div", class_ = "inline-block ratings-imdb-rating").strong.text if general_div.find("div", class_ = "inline-block ratings-imdb-rating") else "")

          stars.append(",".join(general_div.find_all("p")[2].get_text(strip = True).strip("Stars:").split(",")) if general_div.find_all("p")[2].get_text(strip = True).strip("Stars:").split(",") else "")
        except:
          names.append("")

          date_fulls.append("")

          times.append("")

          categories.append("")

          ratings.append("")

          stars.append("")
        #####################################################################################################################################################
        try:
          name_href = first_div.find("a")["href"]
          name_link = 'https://www.imdb.com' + str(name_href) + '?ref_=adv_li_tt'
          
          source2 = requests.get(name_link, headers = headers)
          soup2 = BeautifulSoup(source2.text, "html.parser")
          try:
            episodes.append(soup2.find("span", class_ = "ipc-title__subtext").text if soup2.find("span", class_ = "ipc-title__subtext") else "")
          except:
            episodes.append("")         
          try:
            seasons.append(soup2.find("select", class_ = "ipc-simple-select__input")['aria-label'])
          except:  
            seasons.append("")
        except:
          print("hata")
          episodes.append("")
          seasons.append("")
        #####################################################################################################################################################
        try:
            all_cast_crew = soup2.find("li", class_ = "ipc-metadata-list__item ipc-metadata-list-item--link").a["href"]
            all_cast_crew_link = 'https://www.imdb.com' + str(all_cast_crew)
      
            source3 = requests.get(all_cast_crew_link, headers = headers)
            soup3 = BeautifulSoup(source3.text, "html.parser")
        
            try:
                if soup3.find_all("table", class_ = "simpleTable simpleCreditsTable")[0]:
                    directed_by = soup3.find_all("table", class_ = "simpleTable simpleCreditsTable")[0]
                    directeds_list = directed_by.find_all("td", class_ = "name")
                    names_dir = []
                    for i in directeds_list:
                        names_dir.append(i.a.text.strip())
                    directeds_name = ",".join(names_dir)   
                    directeds.append(directeds_name) # Yönetmen
                else:
                    directeds.append("")
            except:
                directeds.append("")
            try:
                if soup3.find_all("table", class_ = "simpleTable simpleCreditsTable")[1]:
                    writing_credit = soup3.find_all("table", class_ = "simpleTable simpleCreditsTable")[1]
                    writing_list = writing_credit.find_all("td", class_ = "name")
                    writers = []
                    for i in writing_list:
                        writers.append(i.a.text.strip())
                    whole_writers = ",".join(writers)
                    scriptwriters.append(whole_writers) # Senarist
                else:
                    scriptwriters.append("")
            except:
                scriptwriters.append("")
            try:
                if soup3.find_all("table", class_ = "simpleTable simpleCreditsTable")[3]:
                    created_by = soup3.find_all("table", class_ = "simpleTable simpleCreditsTable")[3]
                    created_by_list = created_by.find_all("td", class_ = "name")
                    createds = []
                    for i in created_by_list:
                        createds.append(i.a.text.strip())
                    whole_createds = ",".join(createds)
                    created_bys.append(whole_createds) #series produced by (Yapımcı)
                else:
                    created_bys.append("")
            except:
                created_bys.append("")
            try:
                if soup3.find_all("table", class_ = "simpleTable simpleCreditsTable")[4]:
                    music =  soup3.find_all("table", class_ = "simpleTable simpleCreditsTable")[4]
                    music_list = music.find_all("td", class_ = "name")
                    msc = []
                    for i in music_list:
                        msc.append(i.a.text.strip())
                    whole_musics = ",".join(msc)
                    musics.append(whole_musics)
                else:
                    musics.append("")
            except:
                musics.append("")
        except:
            directeds.append("")
            scriptwriters.append("")
            created_bys.append("")
            musics.append("")
        #####################################################################################################################################################
        try:
            company_credits_href = soup3.find_all('li', class_ = "subnav_item_main")[3].a["href"]
            company_credits_link = "https://www.imdb.com" + str(company_credits_href) + "?ref_=ttco_ql_4"

            source4 = requests.get(company_credits_link, headers = headers)
            soup4 = BeautifulSoup(source4.text, "html.parser")
          
            try:
              if soup4.find_all("ul", class_ = "simpleList")[0]:
                production_companie_list = soup4.find_all("ul", class_ = "simpleList")[0]
                prod_comp_list = production_companie_list.find_all("li")
                prod_comp = []
                for i in prod_comp_list:
                    prod_comp.append(i.a.text.strip())
                whole_prod_comp = ",".join(prod_comp)
                production_companies.append(whole_prod_comp) # Yapım şirketi
              else:
                production_companies.append("")
            except:
              production_companies.append("")
            try:
              if soup4.find_all("ul", class_ = "simpleList")[1]:
                publisher_list = soup4.find_all("ul", class_ = "simpleList")[1]
                pblshr_list = publisher_list.find_all("li")
                pblshr = []
                for i in pblshr_list:
                    pblshr.append(i.a.text.strip())
                whole_publisher = ",".join(pblshr)
                publisher_organization.append(whole_publisher) # Yayıncı kanal
              else:
                publisher_organization.append("") # Yayıncı kanal

            except:
              publisher_organization.append("")
        except:
          production_companies.append("")
          publisher_organization.append("")
        #####################################################################################################################################################

        try:
          release_date_href = soup4.find_all('li', class_ = "subnav_item_main")[1].a["href"]
          realese_date_link = "https://www.imdb.com" + str(release_date_href) + "?ref_=ttco_ql_2"
          source5 = requests.get(realese_date_link, headers = headers)
          soup5 = BeautifulSoup(source5.text, "html.parser")

          try:
            if soup5.find_all("td", class_ = "release-date-item__country-name"):
              exprtd_cntrs = soup5.find_all("td", class_ = "release-date-item__country-name")
              exprts = []
              for i in exprtd_cntrs:
                  exprts.append(i.a.text.strip())
              whole_exports = ",".join(exprts)
              exported_countries.append(whole_exports)
            else:
              exported_countries.append("")
          except:
            exported_countries.append("")
          try:
            release_years.append(soup5.find("td", class_ = "release-date-item__date").text if soup5.find("td", class_ = "release-date-item__date") else "")
          except:
            release_years.append("")
        except:
          exported_countries.append("")
          release_years.append("")

        #####################################################################################################################################################

    starter += 250


# 50 dizi için 4m 29.3s
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": names,
    "Date": date_fulls,
    "Time": times,
    "Category": categories,
    "Rating": ratings,
    "Stars": stars,
    "Episodes": episodes,
    "Seasons": seasons,
    "Directeds": directeds,
    "Scriptwriters": scriptwriters,
    "Release_years": release_years,
    "Production_companies": production_companies,
    "Publisher_organization": publisher_organization,
    "Musics": musics,
    "Created_bys": created_bys,
    "Exported_countries": exported_countries
})

df.index += 1

print(len(df))
df.head()
```
